## Enhancing 14th Century Artwork Through Modern AI Tools

### Introduction
For this project I selected a cultural heritage object from th eUniversity of Glasgow Archives and Special Collections. I decided to choose a collection of letters that had been extracted from a 14th century Caramelite Missal. The object was accessed through the University of Glasgow's digital collections, and the specific item can be found here: ([Source](https://www.gla.ac.uk/collections/#/details?irn=260822&catType=C))

I chose this object due to the small hidden pieces of artwork that is displayed inside each letter.My aim with this project is to highlight the art using artificial intelligence (AI). When first handling the object, I did not see the small illustrations I had to look particularly close in order to be abe to see them, even then, I could not fully see what they were depicting. Therefore, through the use of image enhancing AI I will attempt to enhance the lost details of this medeival manuscript.

### About the Object
A missal is a collection of religeous texts and prayers and was a popular liturgical book dating back to the 5th century ([Source](https://www.britannica.com/topic/missal)). When considering the fact that these letters from this manuscript were once part of a complete missal, it is worth considering the time and skill it took to create such detailed images. Some of the letters in this manuscript are incredeibly small and the illustrations within them are almost invisible at first glance


### Objectives
The main objectives of this portfolio are to explore and demonstrate how AI tools can be applied to cultural heritage objects and assess the potential that AI could have in shedding a new perspective on historical objects, whilst also analysing how the AI interperates prompts and images. Specifically with a focus on this 14th century missal.

A key objective is to explore how AI can enhance or reveal details that the human eye might miss or that the process od time may have damaged. By enhancing these componants this project aims to shed a new perspective on the manuscript and show off its striking visual features and promote a deeper engagement with the object.

In addition to this, this project aims to assess how different AI tools interperate textual prompts and visual inputs. By comparing the output of both AI tools, this project will explore how the input prompt influences the output and how different models respond to the same promts.

Lastly, this project will explore how AI and human collaboraition can impact engagement with heritage objects and whether the output of the AI benefits the material or influences human interpretation.
