## Enhancing 14th Century Artwork Through Modern AI Tools

### Introduction
For this project I selected a cultural heritage object from th eUniversity of Glasgow Archives and Special Collections. I decided to choose a collection of letters that had been extracted from a 14th century Caramelite Missal. The object was accessed through the University of Glasgow's digital collections, and the specific item can be found here: https://www.gla.ac.uk/collections/#/details?irn=260822&catType=C

I chose this object due to the small hidden pieces of artwork that is displayed inside each letter.My aim with this project is to highlight the art using artificial intelligence (AI). When first handling the object, I did not see the small illustrations I had to look particularly close in order to be abe to see them, even then, I could not fully see what they were depicting. Therefore, through the use of image enhancing AI I will attempt to enhance the lost details of this medeival manuscript.

### Objectives
The objectives of this portfolio are to explore and demonstrate how AI tools can be applied to cultural heritage objects and assess the potential that AI could have in shedding a new perspective on historical objects, whilst also analysing how the AI interperates prompts and images.
